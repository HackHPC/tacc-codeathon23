```python
import random

class Person():
    def __init__(self):
        self.current_status = "Susceptible"
        self.days_sick = 0
        self.is_sick = False
        self.is_innoculated = False;
        
        
    def get_status(self):
        return self.current_status
    
    def set_status(self, new_status):
        self.current_status = new_status
        
    def infect(self, number_of_days):
        if (self.is_innoculated == False):
            self.is_sick = True
            self.days_sick = number_of_days
            self.current_status = "Infected"
    
    def update(self):
        if (self.current_status == "Infected"):
            self.days_sick = self.days_sick-1
            
            if (self.days_sick <= 0):
                self.current_status = "Recovered"
                self.is_sick = False
                self.is_innoculated = True

                
```

```python
##Coding Check

joe = Person()
print (joe.get_status())
print (joe.is_sick)
print (joe.is_innoculated)
print (joe.days_sick)

```

```python
joe.infect(120)
print (joe.get_status())
print (joe.is_sick)
print (joe.is_innoculated)
print (joe.days_sick)
```

```python
joe = Person()
max_days_sick = 5
days = 0

while (joe.get_status() != "Recovered"):
    days = days + 1
    roll = random.randint(0,100)
    
    if (roll < 10 and joe.get_status() == "Susceptible"):
        joe.infect(max_days_sick)
    
    print ("Day:", days, "Joe's Status:", joe.get_status())
    
    joe.update()

print ("Day:", days, "Joe's Status:", joe.get_status())

print("Run Complete")
```
